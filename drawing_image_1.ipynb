```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
def plt_imshow(title,image):
  image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
  plt.imshow(image)
  plt.title(title)
  plt.grid(False)
  plt.show()
```

#Implementing Drawing function with OpenCV

```python
# initilize the canvas as 300x300 pixels with 3 channels
#(Red, Green, Blue) with a black background
canvas = np.zeros((300, 300, 3), dtype="uint8")
```

```python
#draw a green line from top left corner of our canvas to the bottom right
green = (0, 255, 0)
cv2.line(canvas, (0, 0), (300, 300), green)
plt_imshow("canvas",canvas)
```

```python
#draw a  3 pixel thick red line from top right corner to the bottom left
red = (0,0,255)
cv2.line(canvas, (300,0), (0,300), red,3)
plt_imshow("canvas", canvas)
```

```python
#draw a green rectangle, this one red with 5 pixel thickness
cv2.rectangle(canvas, (50,200), (200,255), red, 5)
plt_imshow("canvas",canvas)
```

```python
#draw a green 50x50 pixel square, starting at 10x10 and ending at 60x60
cv2.rectangle(canvas,(10,10),(60,60), green,3)
plt_imshow("canvas", canvas)
```

```python
#draw a final rectangle blue filled
blue =(255,0,0)
cv2.rectangle(canvas,(200,50),(225,125),blue,-1)
plt_imshow("canvas",canvas)
```

```python
# re-initialize our canvas as an empty array, then compute the
# center (x, y)-coordinates of the canvas

canvas = np.zeros((300,300,3), dtype="uint8")
(centerX, centerY) = (canvas.shape[1] // 2, canvas.shape[0] // 2)
white = (255,255,255)

#loop over increasing radii, from 25pixels to 150 pixels in 25 pixel increments
for r in range(0,175,25):
  cv2.circle(canvas, (centerX, centerY),r,white)

plt_imshow("canvas",canvas)
```

```python
# re-initialize our canvas once again
canvas = np.zeros((300, 300, 3), dtype="uint8")

# let's draw 25 random circles
for i in range(0, 25):
	# randomly generate a radius size between 5 and 200, generate a
	# random color, and then pick a random point on our canvas where
	# the circle will be drawn
	radius = np.random.randint(5, high=200)
	color = np.random.randint(0, high=256, size=(3,)).tolist()
	pt = np.random.randint(0, high=300, size=(2,))

	# draw our random circle on the canvas
	cv2.circle(canvas, tuple(pt), radius, color, -1)

# display our masterpiece to our screen
plt_imshow("Canvas", canvas)
```
