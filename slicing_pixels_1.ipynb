# Import the Libraries

```python
from matplotlib import pyplot as plt
import argparse
import cv2
```

```python
def plt_imshow(title, image):
  #Covert the image from BGR to RGB colour space and display it
  image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  plt.imshow(image)
  plt.title(title)
  plt.grid(False)
  plt.show()
```

```python
args = {"image" : "/content/drive/MyDrive/CV_course/30th_birthday.png"}
```

```python
image = cv2.imread(args["image"])
```

```python
dimension = image.shape
```

```python
height = image.shape[0]
width = image.shape[1]
channel = image.shape[2]
```

```python
print ('Image Dimension: ', dimension)
print('Image Height: ',height)
print('Image Width: ',width)
print('Number of Channel: ',channel)
```

```python
plt_imshow("Image",image)
```

```python
# images are simply NumPy arrays -- with the origin (0, 0) located at the top-left of the image

(b,g,r) = image[0,0]
print("Pixels at (0,0)- Red: {},Green: {},Blue:{}".format(r,g,b))

#Acess the pixels located at x=50, y= 30
(b,g,r) = image[20,50]
print("Pixels at (50,30) - Red:{}, Green: {}, Blue: {}".format(r,g,b))

#Update the pixels at (50,30) and set it to red
image[20,50] = (0,0,255)
(b,g,r) = image[20,50]
print("Pixel at (50,30) - Red: {}, Green: {}, Blue: {}".format(r,g,b))
```

```python
#Compute the center of the image, which is simply the width & heightdivided by two
(cX, cY) = (width // 2, height // 2)

#Applying array slicing to grab large chunk of reason
t1 = image[0:cY,0:cX]
plt_imshow("Top-Left Corner",t1)
```

```python
#in a similar manner, we can crop the top-right, bottom-right and bottom-left corner of the image
#then display it to our screen

tr = image[0:cY, 0:cX:width]
br = image[cY:height,cX:width]
bl = image[cY:height, 0:cX]

plt_imshow("Top Right Corner", tr)
plt_imshow("Bottom Right Corner", br)
plt_imshow("Bottom Left Corner", bl)
```

```python
#Set the top-right corner of the original image to the screen
image[0:cY, 0:cX] = (0,255,0)

#show out updated image
plt_imshow("Updated", image)
```
